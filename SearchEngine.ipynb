```python
import numpy as np
import pandas as pd
import nltk
import math
import os

from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
```

```python
files_text = []
for i in range(len(files_paths)):
    with open(files_paths[i], 'r') as file: 
        files_text.append(file.read())

files_text
```

```python
def preprocessing(files_text):
    porterStemmer = PorterStemmer()

    corpus = []
    tmp_corpus = []
    for i in range(len(files_text)):
        text = files_text[i].lower() 
        tokens = word_tokenize(text)

        words = []
        for word in tokens:
            stemming = porterStemmer.stem(word)
            words.append(stemming)

        corpus.append(words)
        tmp_corpus.append(' '.join(corpus[i]))

    return corpus, tmp_corpus

corpus = preprocessing(files_text)[0]
tmp_corpus = preprocessing(files_text)[1]
```

```python
def get_unique_values(corpus):
    unique_values = []
    for sublist in corpus:
        for word in sublist:
            if word not in unique_values:
                unique_values.append(word)
    return unique_values
```

```python
unique_values = get_unique_values(corpus)
print(unique_values, '\n')
print(len(unique_values))
```

```python
def getDocID(text):
    text = preprocessing([text])[1][0] # [0][0][0]
    doc_id = []
    for i in range(len(corpus)):
        for word in corpus[i]:
            if text == word:
                doc_id.append(i)
                
    return doc_id
```

```python
getDocID('antony')
```

```python
def getPositions(text, corpus, doc_ids):
    text = preprocessing([text])[0][0][0] # [1][0]
    positions = {}
    for i in doc_ids:
        for j in range(len(corpus[i])):
            if text == corpus[i][j]:
                if i not in positions:
                    positions[i] = [j]
                else:
                    if j not in positions[i]:
                        positions[i].append(j)
    return positions
```

```python
getPositions('antony', corpus, getDocID('antony'))
```

```python
positional_index = {}

for key in unique_values:
    positional_index[key] = [len(getDocID(key)),getPositions(key, corpus, getDocID(key))]

positional_index
```

```python
def phraseQuery(query, positional_index):
    query = query.split()
    query = preprocessing(query)[0]

    query_positions = []
    for i in range(len(query)):
        for word in query[i]:
            try:
                query_positions.append(positional_index[word][1])
            except:
                return f'({word}) -> is not in the dictionary'
    
    query_keys = []
    for i in query_positions:
        query_keys.append(list(i.keys()))
    
    common_doc_id = []
    if query_keys != []:
        for element in set(query_keys[0]):
            flag = True
            for sublist in query_keys[1:]:
                if element not in set(sublist):
                    flag = False
                    break
            if flag:
                common_doc_id.append(element)
                
    common_doc_id.sort()

    final_search = []
    for key in common_doc_id:
        for i in range(len(query_keys) - 1):
            if query_positions[i][key][0] - query_positions[i + 1][key][0] == -1:
                final_search.append(key)
        
    for key in positional_index:
        for inner_key in positional_index[key][1]:
            if len(positional_index[key][1][inner_key]) > 1:
                if inner_key in common_doc_id:
                    try:
                        if query_positions[0][inner_key][0] - query_positions[1][inner_key][1] == -1: 
                            final_search.append(inner_key)
                    except:
                        pass

    if len(query) != 1:
        return list(np.unique(final_search))
    else:
        return query_keys[0]
```

```python
query = 'where fools'
search = phraseQuery(query, positional_index)

if isinstance(search, list):
    for i in search:
        print(f'({query}) -> is found in doc_id ({i})')
else:
    print(search)
```

```python
def booleanPhraseQuery(query, positional_index):
    boolean = ['and', 'or', 'but-not']
    
    for word in boolean:
        if word in query:
            query_parts = query.split(word, 1)
            
            if word == boolean[0]:
                search1 = phraseQuery(query_parts[0], positional_index)
                search2 = phraseQuery(query_parts[1], positional_index)
                result = list(set(search1) & set(search2))
                
            elif word == boolean[1]:
                search1 = phraseQuery(query_parts[0], positional_index)
                search2 = phraseQuery(query_parts[1], positional_index)
                result = list(set(search1) | set(search2))
                
            else:
                search1 = phraseQuery(query_parts[0], positional_index)
                search2 = phraseQuery(query_parts[1], positional_index)
                result = list(set(search1) - set(search2))
            
            query = " ".join(query_parts[1:])
            
    return result
```

```python
def get_weighted_term_freq(tf):
    if tf > 0:
        return math.log10(tf) + 1
    return 0
```

```python
def get_inverse_doc_freq(n, df):
    if df > 0:
        return math.log10(n / df)
    return 0
```

```python
def get_normalized(tf_idf, total_length):
    if total_length > 0 :
        return tf_idf / total_length
    return 0
```

```python
from collections import Counter

def tf_query(query):
    query = query.split()
    query = preprocessing(query)[1]
    
    tf = []
    for word in normlized_tf_idf.index:
        if word in query:
            tf.append(Counter(query)[word])
            # tf.append(query.count(word))
        else:
            tf.append(0)
            
    return tf
```
